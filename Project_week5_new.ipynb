```python
#Cost-Effectiveness Threshold:
#H0: The completion rate for the Test group is less than or equal to the Control group plus 5%.
#H1: The completion rate for the Test group exceeds the Control group plus 5%.
```

```python
completed_test = 18682
completed_control = 15428
total_test = 96159
total_control = 74770

# Completion rates
completion_rate_test = completed_test / total_test
completion_rate_control = completed_control / total_control

# 5% threshold
threshold = 0.05

# Pooled proportion
p_pooled = (completed_test + completed_control) / (total_test + total_control)

# Standard error
se = np.sqrt(p_pooled * (1 - p_pooled) * (1 / total_test + 1 / total_control))

# Z-score for the hypothesis test
z_score = (completion_rate_test - (completion_rate_control + threshold)) / se

# p-value (one-sided test)
p_value = 1 - stats.norm.cdf(z_score)

# Output the z-score and p-value
print(f"Z-Score: {z_score}")
print(f"P-Value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The Test group completion rate exceeds the Control group completion rate plus 5%.")
else:
    print("Fail to reject the null hypothesis: No significant difference in completion rates.")
```

```python
#Avg time during each step in the process
#H0: The avg time spend during each step for both groups are equal.
#H1:  The avg time spend during each step for both groups are not equal.
```

```python
# Chronological order
df_filtered = df_filtered.sort_values(by=["client_id", "date_time"])

# Calculate time difference in seconds between consecutive steps for each client
df_filtered["time_difference"] = df_filtered.groupby("client_id")["date_time"].diff().dt.total_seconds()

# Remove or handle negative time differences 
df_filtered = df_filtered[df_filtered["time_difference"] >= 0]

df_test = df_filtered[df_filtered["variation"] == "Test"]
df_control = df_filtered[df_filtered["variation"] == "Control"]

# Grouping by "step_number" and calculating the mean "time_difference" for each step 
avg_time_test = df_test.groupby("process_step")["time_difference"].mean().round(2)
avg_time_control = df_control.groupby("process_step")["time_difference"].mean().round(2)

avg_time_comparison = pd.DataFrame({
    "Avg_Time_Test": avg_time_test,
    "Avg_Time_Control": avg_time_control
}).dropna()

p_values = []
t_stats = []
steps = avg_time_comparison.index

for step in steps:
       test_times = df_test[df_test["process_step"] == step]["time_difference"].dropna()
    control_times = df_control[df_control["process_step"] == step]["time_difference"].dropna()
    
    # t-test
    t_stat, p_value_ttest = stats.ttest_ind(test_times, control_times, equal_var=False)  
    t_stats.append(round(t_stat,4))
    p_values.append(round(p_value_ttest,4))


avg_time_comparison["T-Statistic"] = t_stats
avg_time_comparison["P-Value"] = p_values


print("Average Time Comparison:")
print(avg_time_comparison)

print("\nT-Test Results:")
for step, t_stat, p_value in zip(steps, t_stats, p_values):
    print(f"Step {step}:")
    print(f"  T-Statistic: {t_stat:.4f}")
    print(f"  P-Value: {p_value:.4f}")  
    
    if p_value < 0.05:
        print(f" Reject the null hypothesis. The average time spent per step is different between the Test and Control groups.")
    else:
        print(f" Fail to reject the null hypothesis. No significant difference in average time spent per step.")
    print() 
```

```python
# Filter rows where process_step is "confirm"
confirmed_df = df_filtered[df_filtered["process_step"] == "confirm"]

new_process_df = confirmed_df[confirmed_df["variation"] == "Test"]
old_process_df = confirmed_df[confirmed_df["variation"] == "Control"]

# Count the number of clients under 50 years old in each group
count_new_process_under_50 = (new_process_df["clnt_age"] < 50).sum()
count_old_process_under_50 = (old_process_df["clnt_age"] < 50).sum()


print(f"Number of clients under 50 in the new process group: {count_new_process_under_50}")
print(f"Number of clients under 50 in the old process group: {count_old_process_under_50}")

# Chi-square test for count comparison

observed_counts = [count_new_process_under_50, count_old_process_under_50]

total_count = count_new_process_under_50 + count_old_process_under_50
expected_proportion = total_count / (len(new_process_df) + len(old_process_df))
expected_counts = [
    expected_proportion * len(new_process_df),
    expected_proportion * len(old_process_df)
]

# Chi-square test
chi2_stat, p_value = stats.chisquare(f_obs=observed_counts, f_exp=expected_counts)

# Results
print(f"Chi-square statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The number of clients under 50 is significantly different between the new and old process.")
else:
    print("Fail to reject the null hypothesis. The number of clients under 50 is not significantly different between the new and old process.")
```
